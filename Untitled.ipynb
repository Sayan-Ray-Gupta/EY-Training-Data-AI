```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Sales dataset
data = {
"Month": ["Jan", "Feb", "Mar", "Jan", "Feb", "Mar", "Jan", "Feb", "Mar"],
"Product": ["Mobile", "Mobile", "Mobile", "Laptop", "Laptop", "Laptop", "TV",
"TV", "TV"],
"Units_Sold": [200, 220, 210, 120, 135, 150, 80, 95, 100],
"Revenue": [300000, 330000, 315000, 600000, 675000, 750000, 240000, 285000,
300000]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
# total revenue per product
total_revenue = df.groupby("Product")["Revenue"].sum()
print (total_revenue)
```

```python
#average units sold per product 
avg_unit = df.groupby("Product")["Units_Sold"].mean()
print(avg_unit)
```

```python
import matplotlib.pyplot as plt 
```

```python
#Visualization 1
plt.figure(figsize=(8,5))
for product in df["Product"].unique():
    subset = df[df["Product"]== product]
    plt.plot(subset["Month"],subset["Revenue"], marker="o", label=product)

plt.title("Monthly Revenue Trend (Jan - Mar)")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.grid(True, linestyle="--", alpha= 0.6)
plt.show()
```

```python
#Visualization 2

plt.figure(figsize=(6,5))
total_revenue.plot(kind="bar", color=["skyblue", "orange", "green"])

plt.title("Total Revenue per Product")
plt.xlabel("Product")
plt.ylabel("Total Revenue")
plt.xticks(rotation=0)
plt.grid(axis="y", linestyle="--", alpha=0.6)

plt.show()
```

```python
#visualisation 3 

plt.figure(figsize=(6,6))
plt.pie(total_revenue, labels=total_revenue.index, autopct='%1.1f%%', startangle= 140, colors=["skyblue", "orange","green"])
plt.title("Contribution of Each Product to Total Revenue")
plt.show()
```
