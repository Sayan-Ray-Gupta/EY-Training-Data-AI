```python
import pandas as pd 
import numpy as np
```

```python
data = {
    "Name": ["Rahul", "Priya", "Arjun", "Neha","Vikram"],
    "Age": [21, 22, 20, 23, 21],
    "Course": ["AI","ML","Data Science","AI","ML"],
    "Marks": [85, 98, 78, 88, 95],
}

df = pd.DataFrame(data)
df
```

```python
print(df["Name"])
```

```python
print(df[["Name","Marks"]])
```

```python
print(df.iloc[0]) #first row
```

```python
print(df.loc[2,"Marks"]) # values at row 2 , column marks
```

```python
#filter data
high_scorer = df[df["Marks"]>85]
print(high_scorer)
```

```python
#add Pass/ Fail columns
df["Result"] = np.where(df["Marks"]>= 80, "Pass", "Fail")
df
```

```python
#Update Neha's Marks
df.loc[df["Name"] == "Neha", "Marks"] = 92
df
```

```python
import matplotlib.pyplot as plt

#group by course and calculate mean
avg_marks = df.groupby("Course")["Marks"].mean()

#Plot bar chart
avg_marks.plot(kind="bar",color="skyblue", edgecolor="red")

plt.title("Average Marks by Course ")
plt.xlabel("course")
plt.ylabel("Average Marks")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
 
# Sample sales dataset
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Product": ["Laptop", "Laptop", "Laptop", "Mobile", "Mobile", "Mobile"],
    "Units_Sold": [120, 135, 150, 200, 220, 210],
    "Revenue": [600000, 675000, 750000, 300000, 330000, 315000]
}
 
df = pd.DataFrame(data)
df
```

```python
#Plot bar chart

units_sold_month = df.groupby("Month")["Units_Sold"].sum()
units_sold_month.plot(kind="bar",color="skyblue", edgecolor="red")
plt.title("Units Sales Report")
plt.xlabel("Units Sold")
plt.ylabel("Months")
plt.xticks(rotation=45)
plt.show()
```
